```python
import pandas as pd

df = pd.read_parquet(DATASET_PATH)
print(df)
```

```python
dataset_size = 10000
print(df[:dataset_size])
```

```python
df = df[:dataset_size]

from collections import Counter

special_tokens = ["<PAD>", "<SOS>", "<EOS>", "<UNK>"]

counter = Counter()

for _, row in df.iterrows():
    user_words = row['user'].lower().split()
    chip2_words = row['chip2'].lower().split()
    counter.update(user_words)
    counter.update(chip2_words)

# Create vocab
words =  [word for word, freq in counter.items() if freq >= 2]  # keep words that appear >=2 times
words = words + special_tokens
word2idx = {word: idx for idx, word in enumerate(words)}
idx2word = {idx: word for word, idx in word2idx.items()}

vocab_size = len(word2idx)

print(f"Vocab size: {vocab_size}")
print(f"Vocab: {word2idx}")
```

```python
def encode(sentence):
    tokens = [word2idx.get(word, word2idx["<UNK>"]) for word in sentence.lower().split()]
    return [word2idx["<SOS>"]] + tokens + [word2idx["<EOS>"]]
```

```python
inputs = []
outputs = []

for _, row in df.iterrows():
    user_input = encode(row['user'])
    chip2_output = encode(row['chip2'])
    inputs.append(user_input)
    outputs.append(chip2_output)

print(inputs[0])
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
import pytorch_lightning as pl
```

$
\begin{align*}
PE_{(pos,\,2i)} &= \sin\left(\frac{pos}{10000^{\frac{2i}{d_{model}}}}\right) \\
PE_{(pos,\,2i+1)} &= \cos\left(\frac{pos}{10000^{\frac{2i}{d_{model}}}}\right)
\end{align*}
$

```python
class PositionEncoding(nn.Module):
    def __init__(self, d_model=2, max_len=6):
        """

        :param d_model - number of the word embedding values per token:
        :param max_len - maximum number of tokens:
        """
        super(PositionEncoding, self).__init__()

        pe = torch.zeros(max_len, d_model)
        position = torch.arange(start=0, end=max_len, dtype=torch.float).unsqueeze(1)

        embedding_index = torch.arange(start=0, end=d_model, step=2, dtype=torch.float)

        div_term = 1/torch.tensor(10000)**(embedding_index / d_model)

        pe[:, 0::2] = torch.sin(position * div_term)
        pe[:, 1::2] = torch.cos(position * div_term)

        """
        tensor([
        [sin, cos]
        ])
        """

        self.register_buffer('pe', pe)

    def forward(self, x):
        """
        Adds position encoding value to word embedding value

        :param:
            x - Word embedding value
        :return:
            out -Word embedding value + position encoding value
        """
        pe = self.pe[:x.size(0)].unsqueeze(1)
        return x + pe
```

$
\begin{align*}
\text{Attention}(Q, K, V) &= \text{SoftMax}\left(\frac{QK^T}{\sqrt{d_k}}\right) * \text{V} \\
\text{Or with mask:} \\
\text{Attention}(Q, K, V) &= \text{SoftMax}\left(\frac{QK^T}{\sqrt{d_k}}\text{+ M}\right) * \text{V} \\
\end{align*}
$

```python
# Head Attention

class Attention(nn.Module):
    def __init__(self, d_model=2):
        """
        :param d_model - number of the word embedding values per token:
        """
        super(Attention, self).__init__()

        # query weights
        self.W_q = nn.Linear(in_features=d_model, out_features=d_model, bias=False)

        # key weights
        self.W_k = nn.Linear(in_features=d_model, out_features=d_model, bias=False)

        # values weights
        self.W_v = nn.Linear(in_features=d_model, out_features=d_model, bias=False)

        self.row_dim = -2
        self.col_dim = -1

    def forward(
        self,
        encodings_for_q,
        encodings_for_k,
        encodings_for_v,
        mask=None
    ):
        """

        :param encodings_for_q (Tensor): Tensor for queries, shape (batch_size, sequence_length, d_model):
        :param encodings_for_k (Tensor): Tensor for keys, shape (batch_size, sequence_length, d_model):
        :param encodings_for_v (Tensor): Tensor for values, shape (batch_size, sequence_length, d_model):
        :param mask (Tensor, optional): Optional mask tensor to prevent attention to certain positions,
                                     shape (batch_size, sequence_length, sequence_length):
        :return:
        """

        q = self.W_q(encodings_for_q)
        k = self.W_k(encodings_for_k)
        v = self.W_v(encodings_for_v)

        # multiply q by the transpose of k. Calculates similarities between the queries and the keys (QK^T)
        sims = torch.matmul(q, k.transpose(dim0=self.row_dim, dim1=self.col_dim))

        # scale the similarities by the square of the number of values used in each key (sqrt(d_k))
        scaled_sims = sims / torch.tensor(k.size(self.col_dim)**0.5)

        if mask is not None:
            # -1e9 is a safe, practical approximation of -inf that avoids numerical instability. These values will be ignored.
            # The mask controls what a token is allowed to "look at" during attention.
            scaled_sims = scaled_sims.masked_fill(mask=mask, value=-1e9)

        attention_percents = F.softmax(scaled_sims, dim=self.col_dim)
        attention_scores = torch.matmul(attention_percents, v)
        return attention_scores
```

```python
# Add and norm / normalization

from torchmetrics import Accuracy

class DecoderTransformer(pl.LightningModule):
    def __init__(self, num_tokens=4, d_model=2, max_len=6):
        super(DecoderTransformer, self).__init__()

        self.we = nn.Embedding(num_embeddings=num_tokens, embedding_dim=d_model)
        self.pe = PositionEncoding(d_model=d_model, max_len=max_len)
        self.attention = Attention(d_model=d_model)
        self.ffn = nn.Sequential(
            nn.Linear(in_features=d_model, out_features=256),
            nn.ReLU(),
            nn.Linear(in_features=256, out_features=num_tokens)
        )
        self.loss = nn.CrossEntropyLoss()

        self.train_accuracy = Accuracy(task="multiclass", num_classes=num_tokens)
        self.val_accuracy = Accuracy(task="multiclass", num_classes=num_tokens)

    def forward(self, token_ids):
        word_embeddings = self.we(token_ids)
        position_encoded = self.pe(word_embeddings)

        seq_len = token_ids.size(1)

        # prevents early tokens from looking at late tokens when calculate Attention
        mask = torch.tril(torch.ones((seq_len, seq_len), device=self.device))

        # converts 0 to True and 1 to False
        mask = mask == 0

        attention_values = self.attention(
            position_encoded,
            position_encoded,
            position_encoded,
            mask=mask
        )

        residual_connection_values = position_encoded * attention_values
        fc_output = self.ffn(residual_connection_values)
        return fc_output

    def configure_optimizers(self):
        return torch.optim.Adam(self.parameters(), lr=0.1)

    def training_step(self, batch, batch_idx):
        inputs, labels = batch
        output = self.forward(inputs)

        output = output.view(-1, output.size(-1))
        labels = labels.view(-1)

        preds = output.argmax(dim=-1)
        loss = self.loss(output, labels)

        self.train_accuracy.update(preds, labels)

        self.log('train_loss', loss)
        self.log("train_acc", self.train_accuracy, prog_bar=True, on_step=False, on_epoch=True)

        return loss

    def validation_step(self, batch, batch_idx):
        inputs, labels = batch
        output = self.forward(inputs)

        output = output.view(-1, output.size(-1))
        labels = labels.view(-1)

        preds = output.argmax(dim=-1)
        loss = self.loss(output, labels)

        self.val_accuracy.update(preds, labels)

        self.log('val_loss', loss)
        self.log("val_acc", self.val_accuracy, prog_bar=True, on_step=False, on_epoch=True)

        return loss
```

```python
class Words2IdxDataset(Dataset):
    def __init__(self, inputs, outputs):
        """
        Args:
            inputs (List[List[int]]): Encoded user inputs
            outputs (List[List[int]]): Encoded chip2 responses
        """
        self.inputs = inputs
        self.outputs = outputs

    def __len__(self):
        return len(self.inputs)

    def __getitem__(self, idx):
        input_ids = torch.tensor(self.inputs[idx], dtype=torch.long)
        output_ids = torch.tensor(self.outputs[idx], dtype=torch.long)
        return input_ids, output_ids
```

```python
from torch.utils.data import random_split

max_input_len = max(len(seq) for seq in inputs)
max_output_len = max(len(seq) for seq in outputs)

max_tokens_len = max_output_len if max_output_len > max_input_len else max_input_len

dataset = Words2IdxDataset(inputs, outputs)

train_size = int(0.8 * len(dataset))
val_size = len(dataset) - train_size

train_dataset, val_dataset = random_split(dataset, [train_size, val_size])
```

```python
from torch.nn.utils.rnn import pad_sequence

batch_size = 32

def collate_fn(batch):
    """
    batch: list of (input_tensor, output_tensor)

    Returns:
        padded_inputs: (batch_size, max_input_len)
        padded_outputs: (batch_size, max_output_len)
    """
    inputs, outputs = zip(*batch)  # separate inputs and outputs

    # Calculate max length for both inputs and outputs
    max_output_len = max([output.size(0) for output in outputs])

    # Pad inputs and outputs to the same max length
    padded_inputs = pad_sequence(inputs, batch_first=True, padding_value=0)
    padded_outputs = pad_sequence(outputs, batch_first=True, padding_value=0)

    # Adjust the padding to make both input and output lengths equal
    if padded_inputs.size(1) < max_output_len:
        padded_inputs = F.pad(padded_inputs, (0, max_output_len - padded_inputs.size(1)))

    return padded_inputs, padded_outputs


train_loader = DataLoader(
    train_dataset,
    batch_size=batch_size,
    shuffle=True,
    collate_fn=collate_fn,
    num_workers=0
)

val_loader = DataLoader(
    val_dataset,
    batch_size=batch_size,
    shuffle=False,
    collate_fn=collate_fn,
    num_workers=0
)
```

```python
from pytorch_lightning.loggers import TensorBoardLogger

logger = TensorBoardLogger("tb_logs", name="my_model")

model = DecoderTransformer(
    num_tokens=vocab_size,
    d_model=max_tokens_len,
    max_len=max_tokens_len
)
trainer = pl.Trainer(
    max_epochs=100,
    logger=logger,
)
trainer.fit(
    model,
    train_dataloaders=train_loader,
    val_dataloaders=val_loader
)
```
